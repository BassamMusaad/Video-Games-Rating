# Import the Required library

```python
import numpy as np
import pandas as pd
import pickle
```

# Import the training set and testing set for tarning the model
##### - The link of dataset is https://www.kaggle.com/imohtn/video-games-rating-by-esrb

```python
training_set = pd.read_csv('dataset/Video_games_esrb_rating.csv')
training_set.head(10)
```

```python
testing_set = pd.read_csv('dataset/test_esrb.csv')
testing_set.head(10)
```

## Check if the traning set and testing set have Nan value

```python
training_set.isnull().sum()
```

```python
testing_set.isnull().sum()
```

# Preprocessing the data

#### - I have deleted the title future from the training set and testing set

```python
training_set.pop('title')
testing_set.pop('title')
```

#### - split the target column from the training set

```python
y = training_set.pop('esrb_rating')
y = y.values
```

###### - take all the remaining column to save them in X_training : will be used to train the model

```python
X = training_set.iloc[:,:].values 
```

#### - do the same thing with testing set

```python
y_testing = testing_set.pop('esrb_rating')
y_testing = y_testing.values
```

###### - take all the remaining column to save them in X_testing : will be used to test the model

```python
X_testing =testing_set.iloc[:,:].values
```

# Import the required library to building the model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score,roc_auc_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

## Hyperparameter tuning
######  - using Grid Search approach

```python
param_grid = {'n_estimators':[50,70,100,150],'criterion':('gini', 'entropy'),"max_depth": [3,5,10, None],
              "min_samples_split": [5,6,7,10],'class_weight':('balanced',None)}
rfc = RandomForestClassifier(random_state=0)
clf = GridSearchCV(rfc, param_grid)
clf.fit(X,y)
clf.best_params_ #return the best paramter that increase the accuracy 
```

## Build the model with optimal hyperparameter

```python
kf = KFold(n_splits=5)
SumAcc=0
fScore=0
RFC = ''
auc=0
SumConf= np.zeros((4,4))
for train, test in kf.split(X):
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    RFC = RandomForestClassifier(criterion='gini',class_weight='balanced',max_depth=None,min_samples_split= 10,n_estimators= 50)
    RFC.fit(X_train,y_train)
    predictions = RFC.predict(X_test)
    SumConf += confusion_matrix(y_test, predictions)
    fScore += f1_score(y_test,predictions,average='weighted')
    SumAcc += accuracy_score(y_test,predictions)
    
print("The Accuracy: " , SumAcc/5) 
print("The Confusion Matrix: " , SumConf/5) 
print("The F1 Score: ", fScore/5)
```

### testing the model

```python
predictions = RFC.predict(X_testing)
SumConf = confusion_matrix(y_testing, predictions)
fScore = f1_score(y_testing,predictions,average='weighted')
SumAcc = accuracy_score(y_testing,predictions)
print("The Accuracy: " , SumAcc) 
print("The Confusion Matrix: " , SumConf) 
print("The F1 Score: ", fScore)
```
